### Etapa 1: Carregamento de dados e preparação dos dados para análise.

#### Padronizar nomes das colunas
A padronização dos dados é uma etapa essencial para garantir consistência, evitar erros nas análises e facilitar operações como merges, agrupamentos e filtragens. No caso deste projeto, um dos principais ajustes é padronizar os nomes das colunas, substituindo letras maiúsculas por minúsculas e adotando um formato uniforme com underscores, removendo espaços. Essa uniformização elimina ambiguidades, previne problemas comuns ao manipular colunas com espaços ou grafias diferentes e torna o código mais limpo e confiável. Além disso, a padronização cria uma base sólida para as etapas seguintes da análise, permitindo que a codificação funcione de forma organizada e sem surpresas.

```python
# Padronizar os nomes das colunas nos arquivos (retirar maísculas e incluir underscores, sem espaçamentos)

visits = visits.rename(columns={
    'Device': 'device',
    'End Ts': 'end_ts',
    'Source Id': 'source_id',
    'Start Ts': 'start_ts',
    'Uid': 'uid'
})

orders = orders.rename(columns={
    'Buy Ts': 'buy_ts',
    'Revenue': 'revenue',
    'Uid': 'uid'
})

costs = costs.rename(columns={
    'source_id': 'source_id',
    'dt': 'dt',
    'costs': 'costs'
})

# Confirmar padronização
print(visits.info())
print(orders.info())
print(costs.info())
```

#### Otimizar os dados e ajustar os tipos de cada coluna
A otimização dos dados e o ajuste correto dos tipos de cada coluna são fundamentais para garantir precisão, eficiência e confiabilidade na análise. Neste projeto, um dos passos mais importantes é a conversão de colunas originalmente armazenadas como 'object' para o tipo 'datetime', especialmente aquelas que representam datas e horários de acesso, início e fim de sessão, compras e despesas. 

Essa conversão permite realizar operações essenciais, como calcular a duração das sessões, agrupar eventos por dia, semana ou mês, analisar o comportamento dos usuários ao longo do tempo e construir métricas como DAU, WAU, MAU e taxa de retenção. Além disso, trabalhar com valores de data no formato adequado reduz erros, melhora o desempenho das consultas e viabiliza análises temporais mais robustas e precisas. Em resumo, ajustar corretamente os tipos das colunas transforma os dados brutos em uma base estruturada, pronta para gerar insights confiáveis.

```python
# Converter os dados com datas e horas para o tipo datetime
visits['start_ts'] = pd.to_datetime(visits['start_ts'])
visits['end_ts'] = pd.to_datetime(visits['end_ts'])

orders['buy_ts'] = pd.to_datetime(orders['buy_ts'])

costs['dt'] = pd.to_datetime(costs['dt'])

# Confirmar conversão
print(visits.info())
print(orders.info())
print(costs.info())
```

```python
# Configurações de estilo
plt.style.use('ggplot')
sns.set_palette('deep')

# Identificar a sessão. Agrupar por usuário e pelo dia para definir a sessão
visits['date'] = visits['start_ts'].dt.date

# Encontrar o tempo mínimo e máximo de atividade para cada usuário e dia
session_duration_df = visits.groupby(['uid', 'date']).agg(
    start_time=('start_ts', 'min'),
    end_time=('start_ts', 'max')
).reset_index()

# Calcular a duração da sessão em segundos
session_duration_df['duration_sec'] = (
    session_duration_df['end_time'] - session_duration_df['start_time']
).dt.total_seconds()

# código revisor
session_duration_df['duration_sec'] = (visits['end_ts'] - visits['start_ts']).dt.seconds

# Limpar outliers: sessões com duração razoável (até 1 hora ou 3600 segundos)
max_duration_seconds = 3600 
session_duration_filtered = session_duration_df.query('duration_sec <= @max_duration_seconds')

# Calcular métricas de centralidade
median_duration = session_duration_filtered['duration_sec'].median()
mean_duration = session_duration_filtered['duration_sec'].mean()
q95_duration = session_duration_filtered['duration_sec'].quantile(0.95)


# Plotar gráfico em barras da duração da sessão 

plt.figure(figsize=(14, 7))

sns.histplot(
    session_duration_filtered['duration_sec'],
    bins=100,
    kde=False,
    color='#3cb371', # Cor verde para destaque
    edgecolor='black' # Bordas pretas para melhor visualização das barras
)

# Adicionar linhas verticais para as métricas de centralidade
plt.axvline(median_duration, color='red', linestyle='--', linewidth=2, label=f'Mediana ({median_duration:.0f}s)')
plt.axvline(mean_duration, color='blue', linestyle=':', linewidth=2, label=f'Média ({mean_duration:.0f}s)')
plt.axvline(q95_duration, color='orange', linestyle='-', linewidth=2, label=f'95º Percentil ({q95_duration:.0f}s)')

plt.title(f'Distribuição do Comprimento das Sessões (Segundos)', fontsize=18, fontweight='bold')
plt.xlabel('Duração da Sessão (Segundos)', fontsize=12)
plt.ylabel('Frequência (Número de Sessões)', fontsize=12)
plt.legend()
plt.xlim(0, max_duration_seconds) # Garante que o gráfico termina em 1 hora
plt.grid(axis='y', linestyle='--', alpha=0.7) # Grade apenas no eixo Y para contagem
plt.tight_layout()
plt.show()
```

```python
# Configurações de estilo
plt.style.use('ggplot')
sns.set_palette('deep')

# Determinar o mês de aquisição de cada usuário (primeira sessão) 
# Criar uma série com o primeiro mês de visita de cada usuário ('uid')
# O método 'transform' garante que essa data seja adicionada a todas as linhas daquele usuário
visits['acquisition_month'] = (
    visits.groupby('uid')['start_ts']
    .transform('min')
    .dt.to_period('M') # Converte para o formato Período (Ano-Mês)
)

# Mês da sessão atual
visits['session_month'] = visits['start_ts'].dt.to_period('M')

# Calculando a "idade" do coorte para cada sessão
# O '.n' extrai o número de meses da diferença de período.
visits['cohort_lifetime'] = (
    visits['session_month'] - visits['acquisition_month']
).apply(lambda x: x.n)

# Tabela de coortes (pivot_table)
cohorts = visits.groupby(
    ['acquisition_month', 'cohort_lifetime']
)['uid'].nunique().reset_index(name='usuarios')

cohorts_pivot = cohorts.pivot_table(
    index='acquisition_month',
    columns='cohort_lifetime',
    values='usuarios'
)

# Guardar o tamanho inicial da coorte (Mês 0) para a normalização
cohort_sizes = cohorts_pivot.iloc[:, 0]

# Dividir cada linha (coorte) pelo seu tamanho inicial e multiplica por 100
retention_pivot = cohorts_pivot.div(cohort_sizes, axis=0) * 100


# Gerar e visualizar o heatmap de retenção

plt.figure(figsize=(16, 9))

sns.heatmap(
    retention_pivot,
    annot=True,     # Mostra o valor percentual dentro de cada célula
    fmt=".1f",      # Formata os valores com uma casa decimal (ex: 25.5)
    cmap="YlGnBu",  # Mapa de cores: Amarelo-Verde-Azul (do menor para o maior valor)
    linewidths=.5,  # Linhas entre as células
    cbar_kws={'label': 'Taxa de Retenção (%)'}
)

# Formatar os rótulos do eixo Y (índice de mês) para melhor visualização
plt.yticks(rotation=0)

plt.title('Taxa de Retenção Mensal por Coorte de Aquisição', fontsize=18, fontweight='bold')
plt.xlabel('Meses de Vida (Lifetime)', fontsize=12)
plt.ylabel('Coorte de Aquisição (Mês da 1ª Visita)', fontsize=12)
plt.tight_layout()
plt.show()
```

### Etapa 3: Análise de vendas. Conversão e frequência de pedidos. Cálculo do Ciclo de Vida do Cliente (CLV - Customer Lifetime Value).

Nesta etapa de Análise de Vendas, **o foco é quantificar o desempenho financeiro dos usuários após o primeiro contato com o produto.** O trabalho começará com a determinação da latência de conversão, **o tempo decorrido entre o primeiro acesso de um usuário e sua primeira compra.** Este é um KPI essencial para avaliar a eficiência dos diferentes canais de marketing e a velocidade com que o usuário se torna um cliente pagante. 

Em seguida, analisarei a **frequência de pedidos e o volume médio de compra (ticket médio) por cliente** para entender o padrão de gasto. 

Finalmente, o ponto alto da análise será o **cálculo do Customer Lifetime Value (CLV), métrica que estima o valor total da receita que cada cliente (ou coorte) gera para a empresa durante seu relacionamento,** fornecendo a base para a otimização das despesas de marketing e a tomada de decisões de investimento.

#### Quando as pessoas começam a comprar o produto?

```python
# Identificar a primeira visita (primeira visita do usuário e contato com o produto)
first_visit = (
    visits.groupby('uid')['start_ts']
    .min()
    .reset_index()
    .rename(columns={'start_ts': 'first_visit_ts'})
)

# Identificar a primeira compra de cada usuário
first_purchase = (
    orders.groupby('uid')['buy_ts']
    .min()
    .reset_index()
    .rename(columns={'buy_ts': 'first_purchase_ts'})
)

# Juntar as duas informações com merge
conversion = first_visit.merge(first_purchase, on='uid', how='inner')

# Calcular o tempo até a conversão (em dias) - 1 dia equivale a 86400 segundos
conversion['days_to_conversion'] = (
    (conversion['first_purchase_ts'] - conversion['first_visit_ts'])
    .dt.total_seconds() / 86400
) 

# Arredondar para facilitar análise por categorias 0d, 1d, 2d, 3d consequentemente
conversion['conversion_day_bucket'] = conversion['days_to_conversion'].round().astype(int)

# Visualizar as primeiras linhas
conversion.head(10)
```

Este tempo até a conversão é um KPI essencial que avalia a **eficácia do produto e das campanhas de marketing.**

Classifiquei os usuários pelo intervalo de conversão. #### Quanto dinheiro eles trazem para a empresa (LTV ou CLV)?

Em análise de negócio, os índices LTV (Lifetime Value) e CLV (Customer Lifetime Value) são essencialmente a mesma coisa e os termos são frequentemente usados. Ambas as métricas se referem ao valor financeiro total que um cliente gera para uma empresa ao longo de todo o seu relacionamento comercial. A variação é mais uma questão de preferência terminológica. **Nesse projeto irei usar a métrica denominada CLV (Customer Lifetime Value), pois é uma sigla mais formal e acadêmica, que especifica claramente que o valor é proveniente do cliente (customer), mas o conceito subjacente é idêntico ao LTV, que é popularmente utilizada em marketing digital e vendas para se referir ao valor vitalício do cliente.**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Encontrar a primeira data de visita para cada usuário (uid)
first_visit_date = visits.groupby('uid')['start_ts'].min()
first_visit_date = first_visit_date.reset_index()
first_visit_date.columns = ['uid', 'cohort_dt']

# Extrair o Mês da Coorte (mês da primeira visita)
first_visit_date['cohort_month'] = first_visit_date['cohort_dt'].dt.to_period('M')

# Juntar os dados de pedidos com a informação da coorte
orders_with_cohort = orders.merge(first_visit_date[['uid', 'cohort_month']], on='uid', how='left')

# Extrair o Mês da Compra
orders_with_cohort['buy_month'] = orders_with_cohort['buy_ts'].dt.to_period('M')


# Calcular a diferença em meses entre o mês da compra e o mês da coorte
orders_with_cohort['cohort_lifetime'] = (orders_with_cohort['buy_month'] - orders_with_cohort['cohort_month']).apply(lambda x: x.n)


# Agrupar a receita e o número de usuários ativos por coorte e lifetime
cohort_revenue = orders_with_cohort.groupby(['cohort_month', 'cohort_lifetime']).agg({
    'revenue': 'sum',
    'uid': 'nunique'
}).reset_index()

# Renomeiar a coluna 'uid' para 'customers' (clientes ativos na coorte)
cohort_revenue.rename(columns={'uid': 'customers'}, inplace=True)


# Calcular o tamanho inicial de cada coorte (número de usuários no mês 0)
cohort_sizes = cohort_revenue.query('cohort_lifetime == 0')[['cohort_month', 'customers']]
cohort_sizes.columns = ['cohort_month', 'cohort_size']

# Juntar o tamanho da coorte com a tabela de receita
clv_table = cohort_revenue.merge(cohort_sizes, on='cohort_month', how='left')

# Calcular a receita média por cliente (CLV - Customer Lifetime Value) para cada mês de vida
clv_table['clv'] = clv_table['revenue'] / clv_table['cohort_size']

# Criar a tabela de CLV cumulativo para visualização
clv_cumulative = clv_table.pivot_table(index='cohort_month', columns='cohort_lifetime', values='clv')
clv_cumulative = clv_cumulative.cumsum(axis=1)
```

#### Cálculo e Análise do Custo de Aquisição de Cliente (CAC)

Para gerar e visulaizar o Custo de Aquisição de Cliente no Python vou realizar as seguintes etapas. Primeiro vou identificar o número de usuários únicos adquiridos (primeira visita) por mês e por origem (source_id). Em seguindo irei cruzar esses dados com os custos mensais por origem (costs). Na sequência vou calcular o CAC (custo / usuários adquiridos).E por fim, irei gerar um heatmap para visualizar a eficiência de cada origem ao longo do tempo.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Encontrar a data e o source_id da primeira visita para cada usuário ('uid')
# Agrupar o DataFrame 'visits' e encontrar o índice da primeira ocorrência de 'start_ts'
idx = visits.groupby('uid')['start_ts'].idxmin()
first_visits = visits.loc[idx].reset_index(drop=True)

# Definir o mês de aquisição
first_visits['acquisition_month'] = first_visits['start_ts'].dt.to_period('M')

# Selecionar e renomeiar as colunas importantes para o CAC 
acquisition_data = first_visits[['uid', 'acquisition_month', 'source_id']]
acquisition_data.columns = ['uid', 'month', 'source_id']

# Contar o número de usuários adquiridos (clientes únicos) por mês e origem
acquired_customers = acquisition_data.groupby(['month', 'source_id'])['uid'].nunique().reset_index()
acquired_customers.columns = ['month', 'source_id', 'new_users']


# Cálculo dos custos mensais

# Agrupar os custos por mês e origem
costs['month'] = costs['dt'].dt.to_period('M')
monthly_costs_by_source = costs.groupby(['month', 'source_id'])['costs'].sum().reset_index()


# Cálculo final do CAC

# Combinar usuários adquiridos e custos
cac_data = pd.merge(
    monthly_costs_by_source,
    acquired_customers,
    on=['month', 'source_id'],
    how='left'
)

# Preencher NaN em 'new_users' com 0 (meses com custo mas sem aquisição no mesmo mês)
cac_data['new_users'] = cac_data['new_users'].fillna(0)

# Calcular o CAC (Custo / Número de Novos Usuários)
# Usar np.where para evitar divisão por zero (CAC será NaN se new_users for 0)
cac_data['cac'] = np.where(
    cac_data['new_users'] > 0,
    cac_data['costs'] / cac_data['new_users'],
    np.nan
)


# Gerar e visualizar o heatmap

# Pivot_table para criar o heatmap do CAC
cac_pivot = cac_data.pivot_table(
    index='source_id',
    columns='month',
    values='cac',
    aggfunc='mean'
)

# Converter o índice de Period('M') para string para melhor visualização no heatmap
cac_pivot.columns = cac_pivot.columns.astype(str)

plt.style.use('ggplot')
plt.figure(figsize=(14, 8))

sns.heatmap(cac_pivot.round(2),
            annot=True,
            fmt=".2f",
            linewidths=.5,
            cmap="Reds", # Mapa de cor 'Reds' para indicar que custos altos são ruins
            cbar_kws={'label':
 'Custo de Aquisição de Cliente (CAC) ($)'}
           )

plt.title('Custo de Aquisição de Cliente (CAC) por Origem e Mês', fontsize=18, fontweight='bold')
plt.ylabel('ID da Origem (Source ID)', fontsize=14)
plt.xlabel('Mês de Aquisição', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()

```

#### Os investimentos valeram a pena? (ROI)

A planilha do ROI (Retorno Sobre o Investimento) a ser trabalhada é a evidência final e mais poderosa da eficiência do investimento em marketing. Irei gerar um heatmap para ilustrar de forma categórica se os investimentos da Y.Afisha valeram a pena ou não.

Na codificação para realização do ROI utilizo a Tabela de Receita por Coorte e Origem, reutilizando a coluna 'orders_with_cohort' da análise CLV, mas adicionando o 'source_id' da primeira visita do usuário.

O 'acquisition_data' foi criado na análise de CAC e contém 'uid', 'month' e 'source_id'. Vou refazer o merge para garantir que tenho o 'source_id' de aquisição associado a cada pedido.

Abaixo explico etapa por etapa para se chegar ao resultado final com a visualização da planilha ROI.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


# Preparar os dados de CLV ('clv_table') e CAC ('cac_data') para cálculo do ROI 

# Aolicar merge 'orders' com o 'source_id' de aquisição
orders_for_roi = orders_with_cohort.merge(
    acquisition_data[['uid', 'source_id']].drop_duplicates(), # pega o source_id da aquisição
    on='uid',
    how='left'
)

# Calcular o lifetime (meses de vida) denominando como 'orders_for_roi'
orders_for_roi['cohort_lifetime'] = (orders_for_roi['buy_month'] - orders_for_roi['cohort_month']).apply(lambda x: x.n)

# Agrupar a receita por coorte, Lifetime e source_id
cohort_revenue_source = orders_for_roi.groupby(
    ['cohort_month', 'cohort_lifetime', 'source_id']
).agg({
    'revenue': 'sum',
    'uid': 'nunique'
}).reset_index()
cohort_revenue_source.rename(columns={'uid': 'customers'}, inplace=True)


# Cálculo de ROI cumulativo

# Cruzar a receita com o custo (CAC)
# O CAC foi calculado na etapa anterior mas preciso do custo total (não dividido) por mês/source
# DataFrame 'monthly_costs_by_source' criado na etapa de CAC.

roi_data = pd.merge(
    cohort_revenue_source,
    monthly_costs_by_source,
    left_on=['cohort_month', 'source_id'],
    right_on=['month', 'source_id'],
    how='left'
)

# Renomeiar as colunas de custo e receita
roi_data.rename(columns={'costs': 'acquisition_cost', 'revenue': 'total_revenue'}, inplace=True)

# Calcular a receita ccumulada por cliente adquirido
# Calcular o tamanho inicial de cada coorte por source_id (clientes adquiridos)
cohort_size_source = acquisition_data.groupby(['month', 'source_id'])['uid'].nunique().reset_index()
cohort_size_source.columns = ['cohort_month', 'source_id', 'cohort_size']

roi_data = roi_data.merge(
    cohort_size_source,
    on=['cohort_month', 'source_id'],
    how='left'
)

# Receita média por cliente (CLV por coorte/source)
roi_data['clv'] = roi_data['total_revenue'] / roi_data['cohort_size']

# Calcular o ROI (CLV Acumulativo / CAC)
# O CAC (Custo por Cliente) é o Custo de Aquisição / Tamanho da Coorte
roi_data['cac'] = roi_data['acquisition_cost'] / roi_data['cohort_size']

# Calcular o ROI em percentual: (Receita Acumulada - CAC) / CAC * 100
# Primeiro, calcula a RECEITA CUMULATIVA por cliente (clv_cumulative)
roi_pivot_clv = roi_data.pivot_table(index=['cohort_month', 'source_id'], columns='cohort_lifetime', values='clv')
roi_pivot_clv = roi_pivot_clv.cumsum(axis=1).reset_index() # Transforma o índice de volta em colunas

# Calcula o ROI (subtrai o CAC do CLV Acumulativo e divide pelo CAC)
# A coluna CAC é constante para cada coorte e origem. Utilizo o CAC do Mês 0.
cac_by_source = roi_data.query('cohort_lifetime == 0')[['cohort_month', 'source_id', 'cac']].drop_duplicates()

# Usar pd.merge em vez de .join() para evitar problemas com MultiIndex
roi_pivot_clv = pd.merge(
    roi_pivot_clv,
    cac_by_source,
    on=['cohort_month', 'source_id'],
    how='left'
)

# Definir o índice novamente para facilitar a subtração
roi_pivot_clv = roi_pivot_clv.set_index(['cohort_month', 'source_id'])

# Aplicar a fórmula do ROI
# Remover a coluna 'cac' temporariamente para a subtração ser correta
clv_columns = roi_pivot_clv.columns.drop('cac')

# Subtrair o CAC (Custo) da Receita Cumulativa (CLV)
roi_numerator = roi_pivot_clv[clv_columns].subtract(roi_pivot_clv['cac'], axis=0)

# Dividir pelo CAC (Custo) e multiplica por 100
roi_pivot = (roi_numerator.div(roi_pivot_clv['cac'], axis=0)) * 100

# Gerar e visualizar o gráfico do ROI
# Filtrar o DataFrame pivotado antes de formatar o índice para string.
top_sources = [1, 2, 3, 4] # as mais relevantes para o exemplo

# Criar um novo DataFrame filtrado usando o método .loc
roi_filtered = roi_pivot.loc[(slice(None), top_sources), :] 
# Filtra as coortes mas apenas os top_sources

# Limpar o eixo de índice para exibição apenas no DataFrame filtrado
roi_filtered.index = roi_filtered.index.map(lambda x: f"Coorte: {x[0].strftime('%Y-%m')} | Origem: {x[1]}")

# Formatar o índice do DataFrame completo (opcional, mas bom para o print final)
roi_pivot.index = roi_pivot.index.map(lambda x: f"Coorte: {x[0].strftime('%Y-%m')} | Origem: {x[1]}")


# Configurações do Plot
plt.style.use('ggplot')
plt.figure(figsize=(15, 10))


sns.heatmap(roi_filtered.round(0), # Plotar APENAS o DataFrame filtrado
            annot=True,
            fmt=".0f",
            linewidths=.5,
            cmap="RdYlGn", # 'Red Yellow Green': Vermelho (Prejuízo), Amarelo (Break-even), Verde (Lucro)
            cbar_kws={'label': 'ROI Cumulativo (%)'}
           )

plt.title('Retorno Sobre o Investimento (ROI) Cumulativo por Coorte e Origem', fontsize=18, fontweight='bold')
plt.ylabel('Coorte de Aquisição e Origem (Source ID)', fontsize=12)
plt.xlabel('Meses de Vida (Lifetime)', fontsize=12)
plt.xticks(rotation=0)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

### Etapa 5: Conclusão e Recomendação Estratégica de Investimento em Marketing

Após analisar profundamente as métricas de CLV (Customer Lifetime Value), CAC (Customer Acquisition Cost) e ROI (Return on Investment) das diferentes origens de tráfego, fica claro que **a Y.Afisha possui um canal de aquisição altamente lucrativo, mas ainda longe de operar no seu potencial máximo devido à distribuição ineficiente do orçamento de marketing.** Em outras palavras: os melhores canais têm espaço para escalar e continuar entregando movimento financeiro saudável, enquanto os mais caros drenam orçamento e precisam ser reavaliados.

A partir dos resultados, proponho uma estratégia direta e sustentável. **Minhas recomendações são redirecionar parte do investimento para os canais mais eficientes e escalar aquilo que já demonstra, de forma consistente, maior retorno por dinheiro investido.**

Minha conclusão final é direta e apoiada pelos dados. É hora de escalar o que funciona e repensar o que drena recursos.

As origens mais eficientes, equivalentes ao que, no mercado, vemos em canais como **Google Search Ads** com intenção alta, campanhas de **retargeting no Meta**, influenciadores de nicho, SEO e conteúdo orgânico, **e-mail marketing automatizado e marketplaces** como Sympla ou Ingressos.com, seguem uma lógica semelhante: baixo custo de aquisição, retorno rápido e alta previsibilidade. São canais que convertem porque **o usuário já chega mais quente**, mais preparado para a compra e **com menor dispersão de atenção**. Esses são exatamente os comportamentos observados nas origens 4, 5, 9 e 10 deste projeto analítico. Por isso, faz sentido **direcionar mais verba para esse grupo:** eles entregam escala com controle e sustentam um crescimento saudável.

Por outro lado, as origens 1, 2 e 3 lembram os canais que, na prática, costumam elevar o CAC sem garantir retorno proporcional: display genérico, campanhas amplas de alcance no Meta, influenciadores grandes demais para o nicho, ou até mídia tradicional quando o produto é digital-first. São investimentos que brilham pela visibilidade, mas nem sempre pelo resultado, e nos dados, esse padrão se repete. **Aqui a recomendação é mitigar, testar e reavaliar, evitando desperdiçar capital em canais que já provaram ter eficiência inferior.**

Por fim, **transformar dados em lucro contínuo exige estratégia, e isso significa operar o marketing como um portfólio vivo.** Ajustes mensais com base no ROI, aumento gradual do orçamento nos canais que superam a meta e cortes inteligentes nos que não entregam; ações de pós-venda para elevar o LTV; testes A/B constantes para ampliar a escala sem perder eficiência; criativos mais orientados à performance para reduzir CAC; expansão por lookalike audiences a partir dos clientes de maior valor; e acompanhamento mensal por coortes para detectar rapidamente quando um canal começa a perder força.

A empresa Y.Afisha tem um enorme potencial de crescimento sustentado. **Com a realocação adequada, ou seja, menos gasto onde dói e mais investimento onde rende, a empresa pode maximizar o retorno do marketing, reduzir risco operacional e fortalecer o ciclo de aquisição de clientes.**

Em resumo, **o caminho para maximizar a lucratividade não é gastar mais, e sim gastar melhor, ampliando os canais mencionados acima que comprovadamente geram valor e ajustando aqueles que não acompanham a performance esperada.** Essa é a base de uma operação de marketing mais inteligente, sustentável e competitiva.
